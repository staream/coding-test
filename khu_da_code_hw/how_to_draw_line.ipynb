```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x_input = np.arange(1,10,2).reshape(-1,1)
y_input = np.array([2,6,10,14,18])
print(x_input)
lr=LinearRegression()
lr.fit(x_input,y_input)
#lr.predict(x_input)
```

```python
x_input1=np.arange(1,100).reshape(-1,1)
pred= lr.predict(x_input1)
print(pred)
plt.scatter(x_input1,pred)
plt.show()
```
